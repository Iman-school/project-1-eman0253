```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('titanic-passengers.csv', sep=';')
df.head(20)
```

```python
#we fill the NaN values of numeric columns with the average value
df['Age'].fillna(df['Age'].mean(), inplace=True)
df.info()
```

```python
df= df.fillna(df.mode().iloc[0])
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Number of missing values in Age
df ['Age'].isnull().sum()
```

```python
#Number of missing values in Cabin
df ['Cabin'].isnull().sum()
```

```python
#Number of missing values in Embarked
df ['Embarked'].isnull().sum()
```

```python
#Delete the cabin column cuz we have so many missing values 
df.drop('Cabin', axis=1, inplace=True)
```

```python
df.info()
```

```python
df = df.dropna() #Drop the rows where at least one element is missing
df.info()
```

```python
#We will create a new feature (column) called family_size by combining ‘SibSp’(Sibling & Spouse) and ‘Parch’(Parent & Children).
def create_fam_size(df):
    return df["SibSp"] + df["Parch"] + 1

df["family_size"] = create_fam_size(df)
```

```python
df.head()
```

```python
import re
def create_categorical_title(df):
    def find_title(name: str) -> str:
        search = re.search(
            " ([A-Za-z]+)\.", name
        )  # Search for a word with a point at the end
        if search:
            title = search.group(1)
            if title in ["Mlle", "Ms"]:
                return "Miss"
            elif title in ["Mme", "Mrs"]:
                return "Mrs"
            elif title == "Mr":
                return "Mr"
            else:
                return "Rare"
        return ""

    return_title = df["Name"].apply(find_title)
    dict_title = {"Miss": 1, "Mrs": 2, "Mr": 3, "Rare": 4}
    return return_title.replace(dict_title)


df["Title"] = create_categorical_title(df)
```

```python
def titanic_children(Passenger):
    
    age , sex = Passenger
    if age <16:
        return 'child'
    else:
        return sex
df['person'] = df[['Age','Sex']].apply(titanic_children,axis=1)
df["person"].value_counts() #Get a number of male ond female
```

```python
df["Alone"] = df.SibSp + df.Parch
```

```python
# if Alone value is >0 then they are with family else they are Alone
df['Alone'].loc[df['Alone']>0] = 'With Family'
df['Alone'].loc[df['Alone']==0] = 'Without Family'
```

```python
df["Survived"].value_counts() #Get a number of who is survived or not
```

```python
cleanup_Survived = {"Survived":{"No": 0, "Yes": 1}}
df = df.replace(cleanup_Survived) #replace Yes & No with 0 & 1
df.head(5)#print the first 5 rows of the data frame
```

```python
cleanup_Sex = {"person" : {"female" : 0, "male": 1, "child": 3}}
df = df.replace(cleanup_Sex) 
df.head()
```

```python
df["person"].value_counts() #Get a number of male and female
```

```python
df["Embarked"].value_counts()
```

```python
cleanup_Embarked = {"Embarked": {"S": 1, "C": 2, "Q":3}}
df = df.replace(cleanup_Embarked)
df.head(20)
```

```python
df["Alone"].value_counts()
```

```python
cleanup_Alone = {"Alone": {"With Family": 1, "Without Family": 0}}
df = df.replace(cleanup_Alone)
df.head(20)
```

```python
# Delete the name column
df.drop("Name", axis=1, inplace= True)
```

```python
# Delete the Ticket column
df.drop("Ticket", axis=1, inplace= True)
```

```python
# Delete the Fare column
df.drop("Fare", axis=1, inplace= True)
```

```python
# Delete the PassengerId column
df.drop("PassengerId", axis=1, inplace= True)
```
